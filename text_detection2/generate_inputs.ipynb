```python
from tensorflow.keras import Model, Input
import cv2
import os
from keras.preprocessing.sequence import pad_sequences
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('../csv/combined-training.csv')
# df = pd.read_csv('../csv/combined-training.csv')
df_col_names = df.columns
df = df.fillna(-1)
```

```python
imgs_dir = '../new_data'
dirs_ = []
for roots, dirs, files in os.walk(imgs_dir):
    dirs_ += dirs

# dirs_.remove('.DS_Store')
```

```python
MAX_SEQ = 100
WIDTH = 48
HEIGHT = 48

X1 = []
X2 = []
X3 = []
y = []
print(len(dirs_))
for i in range(len(dirs_)):
    for roots__, dirs__, files__ in os.walk(os.path.join(imgs_dir, dirs_[i])):
        fs_ = files__
    print(i)
    print("Current Dir:{}".format(dirs_[i]))
    n_patches = len(fs_) - 1
    if n_patches>0:
        merged_patches = []
        
        for j in range(n_patches):
            patch_path = os.path.join(imgs_dir, dirs_[i], dirs_[i]+'_{}.jpg'.format(j))
            img_ = cv2.imread(patch_path)
            img_ = cv2.resize(img_, (WIDTH, HEIGHT))
            merged_patches.append(img_)
        
        merged_patches = np.concatenate(merged_patches, axis=1)
        merged_patches = pad_sequences(sequences=merged_patches, maxlen=MAX_SEQ*WIDTH, padding='post')
        
        bbox = np.load(os.path.join(imgs_dir, dirs_[i], dirs_[i]+'_bbox.npy'))
        bbox = bbox.reshape((bbox.shape[0], 8))
        bbox = np.einsum('ji->ij', bbox)
        bbox = pad_sequences(sequences=bbox, maxlen=MAX_SEQ, padding='post')
        
        row_ = df[df['Image-name']==dirs_[i]+'.jpg']
        target_bbox = [row_[df_col_names[2+i*10:10+i*10]].values for i in range(7)]
        
        img_m = cv2.imread(os.path.join('../cutted_data/', dirs_[i])+'.jpg')
        img_m = cv2.resize(img_m, (299,299))
        
        X1.append(merged_patches)
        X2.append(bbox)
        X3.append(img_m)
        y.append(np.squeeze(target_bbox, axis=1))
```

```python
X1 = np.array(X1)
X2 = np.array(X2)
X3 = np.array(X3)
y = np.array(y)
```
